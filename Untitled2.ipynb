```python
import cv2
import random
from ultralytics import YOLO
from IPython.display import clear_output
from google.colab.patches import cv2_imshow

# Read class names from file
with open('/content/cock.txt', 'r') as my_file:
    class_list = my_file.read().split('\n')

# Generate random colors
detection_color = []
for _ in range(len(class_list)):
    r = random.randint(0, 255)
    g = random.randint(0, 255)
    b = random.randint(0, 255)
    detection_color.append((r, g, b))

# Load model
model = YOLO('yolov8n.pt')

# Open video file
cap = cv2.VideoCapture('/content/126434-735976920_small.mp4')
if not cap.isOpened():
    print("We cannot open file")
    exit()

# Loop through video frames
while True:
    ret, frame = cap.read()
    if not ret:
        print("Finished processing video.")
        break

    # Perform detection
    detect_params = model.predict(source=[frame], conf=0.45, save=False)
    boxes = detect_params[0].boxes

    for box in boxes:
        clsID = int(box.cls.cpu().numpy()[0])
        conf = box.conf.cpu().numpy()[0]
        bb = box.xyxy.cpu().numpy()[0]

        # Draw rectangle
        cv2.rectangle(frame, (int(bb[0]), int(bb[1])), (int(bb[2]), int(bb[3])),
                      detection_color[clsID], 2)

        # Add label
        label = f'{class_list[clsID]} {round(conf * 100, 2)}%'
        cv2.putText(frame, label, (int(bb[0]), int(bb[1]) - 10),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2)

    # Display the frame
    clear_output(wait=True)
    cv2_imshow(frame)
    cv2.waitKey(1)

# Release resources
cap.release()
cv2.destroyAllWindows()
```
